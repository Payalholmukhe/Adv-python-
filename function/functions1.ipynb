```python
isinstance(34 , int)
```

```python
isinstance(56.89,(int,float))
```

```python
def add(x:int|float , y :int|float): 
    """ this is used to add two number only """
    assert isinstance(x,(int,float)),"please insert x as int or float"
    assert isinstance(y,(int,float)),"please insert y as int or float"
    z = x+y
    return z




```

```python

def add(x ,y):                                  #alternate 
    if isinstance(x, (int , float) ) and  isinstance(y , (int , float)):
        z =  x+y
        return z 
    else:
        print('please insert int or float type of data') 
```

```python
add(55,55)
```

```python
add(y=23 , x=45 )       #keyword argument


```

```python
def funtion(*args):
    total = 0
    for i in args:
        total = total+i
    return total

    

```
